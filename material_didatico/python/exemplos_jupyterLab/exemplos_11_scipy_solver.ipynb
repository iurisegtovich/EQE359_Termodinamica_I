```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def f(x):
    return np.log(x)

goal=2.

def res(x):
    return f(x)-goal
```

```python
x=np.linspace(1,10,100)
```

```python
y=f(x)
```

```python
plt.plot(x,y)
plt.axhline(y=goal)
```

```python
from scipy import optimize as opt
xsol=opt.bisect(f=res,a=2.,b=10)
```

```python
plt.plot(x,y)
plt.axhline(0)
plt.scatter(xsol,res(xsol),c='r')
```

# salvando a historia

```python
histx=[]
histf=[]

def f(x):
    histx.append(x)
    fx=np.log(x)
    histf.append(fx)
    return fx

goal=2.

def res(x):
    return f(x)-goal

from scipy import optimize as opt
xsol=opt.bisect(f=res,a=2.,b=10)

plt.plot(x,y)
plt.axhline(0)
plt.scatter(histx,histf,c='y')
plt.scatter(xsol,f(xsol),c='r')
```
