```python
import numpy as np
```

```python
def f(x):
    return np.log(x)-2.
```

```python
x=np.linspace(1,10,100)
```

```python
y=f(x)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.plot(x,y)
plt.axhline(y=0)
```

```python
from scipy import optimize as opt
xsol=opt.fsolve(func=f,x0=2.)
```

```python
plt.plot(x,y)
plt.axhline(0)
plt.scatter(xsol,f(xsol),c='r')
```
