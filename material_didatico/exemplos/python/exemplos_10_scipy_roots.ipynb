```python
import numpy as np
from scipy import roots
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
a=12
b=.34
c=5.9
d=3.7

def y(x):
    return a*x**3+b*x**2+c*x+d

x=np.linspace(-1,1,1000)

plt.plot(x,y(x))
plt.axhline(0)

raizes=roots((a,b,c,d))

for raiz in raizes:
    if np.isreal(raiz):
        plt.scatter(raiz.real,y(raiz.real),color='r')
```
