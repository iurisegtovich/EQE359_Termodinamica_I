```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
a=32
b=3.34
c=-20
d=3.7

def y(x):
    return a*x**3+b*x**2+c*x+d

x=np.linspace(-1,1,1000)
```

```python
plt.plot(x,y(x))
plt.axhline(0)
```

```python
raizes=np.roots((a,b,c,d))
```

```python
plt.plot(x,y(x))
plt.axhline(0)
for raiz in raizes:
    if np.isreal(raiz):
        plt.scatter(raiz.real,y(raiz.real),color='r')
```
